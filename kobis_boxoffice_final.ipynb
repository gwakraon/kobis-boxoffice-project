```python
# ✅ 1. 라이브러리 불러오기
import requests
import pandas as pd
from datetime import datetime, timedelta

# ✅ 2. 오늘 날짜 전날로 맞추기 (KOBIS는 하루 전 데이터 제공)
yesterday = (datetime.today() - timedelta(days=1)).strftime('%Y%m%d')

# ✅ 3. API 키 입력 (발급받은 값으로 바꾸기)
API_KEY = "69454813a1278e53a6984dbe05fcf70f"

# ✅ 4. API 요청 URL 구성
url = f"http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key={API_KEY}&targetDt={yesterday}"

# ✅ 5. 요청 및 응답 처리
res = requests.get(url)
data = res.json()
movies = data['boxOfficeResult']['dailyBoxOfficeList']

# ✅ 6. 필요한 정보 추출
movie_data = []
for m in movies:
    movie_data.append({
        '순위': m['rank'],
        '영화명': m['movieNm'],
        '개봉일': m['openDt'],
        '당일 관객수': m['audiCnt'],
        '누적 관객수': m['audiAcc']
    })

# ✅ 7. DataFrame으로 정리
df = pd.DataFrame(movie_data)

# ✅ 8. 출력 및 저장
display(df)
df.to_excel("박스오피스_오늘.xlsx", index=False)
print("✅ 박스오피스 정보 저장 완료!")
```
